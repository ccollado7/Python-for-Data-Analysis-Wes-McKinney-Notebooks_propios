# Lineal Algebra

```python
import numpy as np
import pandas as pd
from numpy.random import randn
```

Algebra lineal, tal como multiplicacion de matrices, desomposicion, determinantes y otras operaciones matematicas sobre matrices cuadradas, son una importante parte de cualquier libreria.

Salvo algunos lenguajes tal como MATLAB, la multiplicacion de dos matrices 2D con el simbolo * es una multiplicacion elemento a elemento y no el producto matricial. Para esta operacion de producto matricial existe una funcion dot (puede ser utilizada como metodo o funcion):

```python
#Creo 2 arrays (2x3) y (3x2) de forma tal que sea posible el producto matricial

x = np.array([[1,2,3],[4,5,6]],dtype=np.float64)
y = np.array([[6,23],[-1,7],[8,9]],dtype=np.float64)
```

```python
x
```

```python
y
```

```python
x.dot(y) #Realizo el producto matricial entre la matriz x e y. Obtengo una matriz 2x2
```

```python
np.dot(x,y) #El mismo resultado,otra sintaxis
```

```python
y.dot(x) #Realizo el producto matricial entre la matriz y e x. Obtengo una matriz 3x3
```

```python
np.dot(y,x)
```

El producto matricial entre un array 2D y un array 1D resulta en un array 1D:

```python
np.dot(x,np.ones(3)) # (2x3) dot (1x3)
```

numpy.linalg tiene un set estandar de descomposicion de matrices, inversa y determinante

```python
from numpy.linalg import inv,qr
```

```python
X = randn(5,5) #Genero un array 5x5 de numeros aleatorios
```

```python
X #Visualizo el array
```

```python
X.T #Traspuesta del array generado anteriormente
```

```python
mat = (X.T).dot(X) #Realizo el producto matricial entre la traspuesta de X y X propiamente dicha
```

```python
inv(mat) #Calculo la inversa de la matriz
```

```python
mat.dot(inv(mat)) #Realizo el producto matricial entre mat y la inversa de mat, por lo tanto debo obtener la matriz identidad (todos 1 en la diagonal principal)
```

```python
q,r = qr(mat)
```

```python
q
```

```python
r
```

### Ejemplo de resolucion de un sistema lineal de 2 ecuaciones A*x = b

```python
A = randn(2,2) #Creo un array 2x2 de numeros aleatorios como matriz de coeficientes A
```

```python
A
```

```python
b = randn(2,1) #Creo un array 2x1 de numeros aleatorios como vector de valores b
```

```python
b
```

```python
np.linalg.solve(A,b) #Resuelvo el sistema de ecuaciones
```
