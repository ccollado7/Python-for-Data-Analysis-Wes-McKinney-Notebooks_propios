# Random Number Generation 

```python
import numpy as np
import pandas as pd
```

El modulo numpy.random suplementa el modulo de Python random con funciones para la generacion eficiente de arrays desde diferentes tipos de distribuciones de probabilidad

Por ejemplo podemos tener un array 4x4 de numeros de la distribucion estandar normal utilizando normal:

```python
sample = np.random.normal(size=(4,4))
```

```python
sample
```

El modulo random de Python, en contraste, solamente da como muestra un valor a la vez.

A continuacion se muestra una comparacion entre numpy.random y ramdom, lo cual muestra que la funcion de NumPy es un orden de magnitud mas rapido para la generacion de muy largos ejemplos:

```python
from random import normalvariate
```

```python
N = 1000000
```

```python
%timeit samples = [normalvariate(0,1) for _ in range(N)]
```

```python
%timeit np.random.normal(size=N)
```

Como se observa anteriormente el resultado utilizando NumPy es mas rapido para tamaños de ejemplos muy grandes, 1 millon en este caso
